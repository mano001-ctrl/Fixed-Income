```python
import pandas as pd

# Sample data: South Africa's 10 most liquid bonds
data = {
    'Bond': ['R186', 'R2030', 'R2032', 'R2044', 'R2048', 'R2023', 'R2026', 'I2025', 'I2038', 'R2053'],
    'Maturity': ['2026', '2030', '2032', '2044', '2048', '2023', '2026', '2025', '2038', '2053'],
    'Yield (%)': [10.5, 11.0, 11.2, 11.3, 11.4, 10.3, 10.6, 9.5, 9.8, 10.7],
    'Duration (years)': [2, 6, 8, 20, 24, 1, 2, 1, 14, 29],
    'Coupon Rate (%)': [10.5, 8.00, 8.25, 8.75, 8.75, 7.75, 8.00, 'Variable', 'Variable', 8.50]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Analyze the data
average_yield = df['Yield (%)'].mean()
max_yield = df['Yield (%)'].max()
min_yield = df['Yield (%)'].min()
median_yield = df['Yield (%)'].median()

average_duration = df['Duration (years)'].mean()

# Display the results
print("Analysis of South Africa's 10 Most Liquid Bonds")
print("-" * 50)
print(f"Average Yield: {average_yield:.2f}%")
print(f"Maximum Yield: {max_yield:.2f}% (Bond: {df[df['Yield (%)'] == max_yield]['Bond'].values[0]})")
print(f"Minimum Yield: {min_yield:.2f}% (Bond: {df[df['Yield (%)'] == min_yield]['Bond'].values[0]})")
print(f"Median Yield: {median_yield:.2f}%")
print(f"Average Duration: {average_duration:.2f} years")
print("-" * 50)

# Optional: Display the DataFrame
print("\nDetailed Bond Information:")
print(df)

# Optional: Save the DataFrame to a CSV file
df.to_csv('south_africa_liquid_bonds_analysis.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: South Africa's 10 most liquid bonds
data = {
    'Bond': ['R186', 'R2030', 'R2032', 'R2044', 'R2048', 'R2023', 'R2026', 'I2025', 'I2038', 'R2053'],
    'Maturity': ['2026', '2030', '2032', '2044', '2048', '2023', '2026', '2025', '2038', '2053'],
    'Yield (%)': [10.5, 11.0, 11.2, 11.3, 11.4, 10.3, 10.6, 9.5, 9.8, 10.7],
    'Duration (years)': [2, 6, 8, 20, 24, 1, 2, 1, 14, 29],
    'Coupon Rate (%)': [10.5, 8.00, 8.25, 8.75, 8.75, 7.75, 8.00, 9.5, 9.8, 8.50]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Plotting the Yield and Coupon Rates
plt.figure(figsize=(10, 6))
plt.bar(df['Bond'], df['Yield (%)'], color='blue', alpha=0.7, label='Yield (%)')
plt.bar(df['Bond'], df['Coupon Rate (%)'], color='orange', alpha=0.7, label='Coupon Rate (%)', width=0.4)
plt.xlabel('Bond')
plt.ylabel('Percentage (%)')
plt.title('Yield and Coupon Rate of South Africa\'s Most Liquid Bonds')
plt.legend()
plt.show()

# Plotting the Duration
plt.figure(figsize=(10, 6))
plt.bar(df['Bond'], df['Duration (years)'], color='green')
plt.xlabel('Bond')
plt.ylabel('Duration (years)')
plt.title('Duration of South Africa\'s Most Liquid Bonds')
plt.show()
```

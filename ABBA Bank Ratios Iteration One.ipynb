```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Function to clean and extract matrix rows for bank ratios
def clean_matrix_rows_v2_bank(matrix_rows):
    cleaned_matrix = []
    
    # Loop through each captured row, clean, and convert to float values
    for row in matrix_rows:
        # Remove square brackets and clean spaces
        cleaned_row = row.replace('[', '').replace(']', '').strip()
        # Convert the cleaned row into a list of floats
        cleaned_matrix.append(list(map(float, cleaned_row.split())))
    
    # Convert to a pandas DataFrame
    return pd.DataFrame(cleaned_matrix)

# Load the bank ratios file (replace with your file path if needed)
bank_ratios_path = 'bank-ratios.csv'

# Read the raw content of the bank ratios file
with open(bank_ratios_path, 'r') as file:
    raw_bank_data = file.read()

# Extract matrix rows using regex for bank ratios
bank_matrix_rows = re.findall(r'\[\s*(.*?)\s*\]', raw_bank_data)

# Clean the matrix rows with the improved function for bank ratios
cleaned_bank_ratios_df = clean_matrix_rows_v2_bank(bank_matrix_rows)

# Display the cleaned bank ratios dataframe for inspection
print("Cleaned Bank Ratios Data:")
print(cleaned_bank_ratios_df.head())  # Show the first few rows of the DataFrame

# Plotting key metrics from the cleaned bank ratios data
plt.figure(figsize=(12, 8))

# Capital Ratio trend
plt.subplot(2, 2, 1)
plt.plot(cleaned_bank_ratios_df.index, cleaned_bank_ratios_df[2], label='Capital Ratio', color='blue')
plt.title('Capital Ratio Trend')
plt.xlabel('Iteration')
plt.ylabel('Capital Ratio')

# Reserves Ratio trend
plt.subplot(2, 2, 2)
plt.plot(cleaned_bank_ratios_df.index, cleaned_bank_ratios_df[3], label='Reserves Ratio', color='green')
plt.title('Reserves Ratio Trend')
plt.xlabel('Iteration')
plt.ylabel('Reserves Ratio')

# Leverage Ratio trend
plt.subplot(2, 2, 3)
plt.plot(cleaned_bank_ratios_df.index, cleaned_bank_ratios_df[4], label='Leverage Ratio', color='red')
plt.title('Leverage Ratio Trend')
plt.xlabel('Iteration')
plt.ylabel('Leverage Ratio')

# Total Assets trend
plt.subplot(2, 2, 4)
plt.plot(cleaned_bank_ratios_df.index, cleaned_bank_ratios_df[13], label='Total Assets', color='purple')
plt.title('Total Assets Trend')
plt.xlabel('Iteration')
plt.ylabel('Total Assets')

plt.tight_layout()
plt.show()
```

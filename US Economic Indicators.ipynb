```python
# Get the 2-Year and 10-Year Treasury yields
two_year_yield = fred.get_series('DGS2')
ten_year_yield = fred.get_series('DGS10')

# Calculate the spread
spread = ten_year_yield - two_year_yield

# Plot the spread
plt.figure(figsize=(10, 6))
spread.plot(title="10-Year vs. 2-Year Treasury Yield Spread", label='Yield Spread')

# Shade the periods where the yield curve is inverted (spread < 0)
inverted_periods = spread[spread < 0]
plt.fill_between(inverted_periods.index, inverted_periods, 0, color='red', alpha=0.3)

# Adding labels and grid
plt.xlabel('Year')
plt.ylabel('Spread (%)')
plt.grid(True)

plt.show()
```

Description of the Chart:
Vertical Axis (Y-Axis): Represents the yield spread between the 10-year and 2-year U.S. Treasury bonds, measured in percentage points (%).
Horizontal Axis (X-Axis): Represents time, spanning from around 1970 to the present.
Key Observations:
Positive Spread (Above 0%):

For much of the time period shown, the yield spread is positive, meaning that the 10-year Treasury yield is higher than the 2-year Treasury yield. This is typical in a normal yield curve scenario, where longer-term bonds yield more due to higher risks and opportunity costs associated with longer time horizons.
Inverted Spread (Below 0%):

There are distinct periods where the spread drops below 0%, indicating an inverted yield curve. An inverted yield curve occurs when short-term yields exceed long-term yields. This phenomenon is widely regarded as a predictor of economic recessions.

The chart shows notable inversions around the following periods:

Late 1970s to Early 1980s: Corresponding to the stagflation period and the severe recessions that followed.

Late 1980s to Early 1990s: Leading up to the early 1990s recession.

Early 2000s: Preceding the dot-com bubble burst and the early 2000s recession.
Mid-2000s: Prior to the Global Financial Crisis of 2008.

Late 2010s: Prior to the economic slowdown and the onset of the COVID-19 pandemic in early 2020.
Recent Trends:

The chart shows a sharp decline in the spread leading into the late 2010s, followed by a recovery and then another decline in recent years.

The recent inversion is associated with the economic uncertainty surrounding the COVID-19 pandemic and concerns about potential future recessions.


Economic Interpretation:
Yield Curve Inversions: The inversion of the yield curve, where the spread turns negative, has historically been a reliable indicator of upcoming recessions. When the market expects economic conditions to deteriorate, investors seek the safety of long-term bonds, driving down their yields relative to short-term bonds, which often leads to an inversion.
Current Spread: The chart’s latest data suggests a narrowing of the spread or a possible inversion, reflecting ongoing concerns in the market about economic stability and growth prospects.

Conclusion:
This chart provides valuable insight into the relationship between short-term and long-term interest rates over time. It highlights how shifts in the yield curve spread can signal changes in economic conditions, particularly the likelihood of a recession. The pattern of yield curve inversions seen in this chart is a critical tool for economists and investors alike, helping to anticipate and prepare for economic downturns.









```python
fed_funds_rate = fred.get_series('FEDFUNDS')
ten_year_yield = fred.get_series('DGS10')

plt.figure(figsize=(10, 6))
fed_funds_rate.plot(label='Federal Funds Rate')
ten_year_yield.plot(label='10-Year Treasury Yield')
plt.title('Federal Funds Rate vs. 10-Year Treasury Yield')
plt.xlabel('Year')
plt.ylabel('Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

Chart Description:
Vertical Axis (Y-Axis): Represents the interest rates (%) for both the Federal Funds Rate and the 10-Year Treasury Yield.

Horizontal Axis (X-Axis): Represents time, spanning from approximately 1955 to 2023.
Key Observations:
Federal Funds Rate (Blue Line):

The Federal Funds Rate is the interest rate at which banks lend reserves to each other overnight. It is a key tool used by the Federal Reserve to control monetary policy.
The chart shows several significant peaks in the Federal Funds Rate:

Late 1970s to Early 1980s: The most prominent peak occurs during the early 1980s, when the Federal Reserve, under Chairman Paul Volcker, raised rates sharply to combat the high inflation of the period. Rates reached nearly 20% during this time.

1990s: The rate was generally declining, with periodic increases to manage economic growth and inflation.

2000s to Present: The rate was significantly lowered during the 2008 financial crisis and remained near zero for an extended period. There were gradual increases starting around 2015, followed by a sharp drop again during the COVID-19 pandemic in 2020.
10-Year Treasury Yield (Orange Line):

The 10-Year Treasury Yield represents the return on investment for U.S. government bonds with a 10-year maturity. It is often considered a benchmark for long-term interest rates.

The yield tends to move in a similar direction to the Federal Funds Rate, though it generally remains higher, reflecting the longer maturity and associated risk premium.

There is a noticeable trend of decreasing yields over the long term, from highs of around 15% in the early 1980s to below 2% in the years following the 2008 financial crisis and during the COVID-19 pandemic.
Relationship Between the Two Rates:

Co-movement: Both rates generally move in tandem, with the Federal Funds Rate often leading changes in the 10-Year Treasury Yield. This relationship underscores the influence of short-term monetary policy on long-term interest rates.

Periods of Divergence: There are instances where the spread between the two widens or narrows, reflecting differing market expectations for short-term vs. long-term economic conditions. For example, the 10-Year Yield did not rise as much as the Federal Funds Rate in the early 2000s, signaling expectations of lower inflation and growth in the future.
Economic Implications:

Policy Tightening: Periods of sharp increases in the Federal Funds Rate typically correspond to efforts by the Federal Reserve to curb inflation, often leading to higher short-term rates and potentially triggering recessions.
Easing Policies: Lowering the Federal Funds Rate, as seen during and after the 2008 financial crisis and the COVID-19 pandemic, is aimed at stimulating the economy by making borrowing cheaper, which often leads to lower yields across the curve.

Conclusion:
This chart effectively demonstrates the central role of the Federal Funds Rate in shaping broader interest rate trends, including the 10-Year Treasury Yield. It highlights how central bank policy decisions impact the yield curve, influencing everything from consumer loans to government borrowing costs. The long-term trend of declining yields, despite periodic spikes in the Federal Funds Rate, reflects broader changes in the economic environment, including lower inflation expectations and slower growth.
